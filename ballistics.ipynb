# Ballistics and the Range Equation
This is an investigation of the so-called *range equation*: $d=\frac {v^2} g \sin 2\theta$.

```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    distance = (v**2/9.81)*sin(2*theta)
    return(distance)
```

```python
angledegrees = linspace(0, 90, num = 200)
rangefromdegrees = [rangeformula(100, radians(angle)) for angle in angledegrees]
print(rangefromdegrees)
```

```python
plt.plot(angledegrees, rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
rangeforangle100 = [rangeformula(100, radians(angle)) for angle in angledegrees]
rangeforangle200 = [rangeformula(200, radians(angle)) for angle in angledegrees]
```

```python
plt.plot(angledegrees, rangeforangle100, label = "100 m/s")
plt.legend()
plt.plot(angledegrees, rangeforangle200, label = "200 m/s")
plt.legend()
plt.title('Range as a function of angle at various velocities')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
angleradians = linspace(0, pi, num = 200)
rangefromradians = [rangeformula(100, angle) for angle in angleradians]
plt.plot(angleradians, rangefromradians)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (radians)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
initialvelocity = linspace(0, 200, num = 200)
rangefromiv = [rangeformula(velocity, radians(45)) for velocity in initialvelocity]
plt.plot(initialvelocity, rangefromiv)
plt.title('Range as a function of initial velocity at 45 degrees')
plt.xlabel('Initial Velocity (m/s)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
initialvelocity = linspace(0, 200, num = 200)
rangefromiv15 = [rangeformula(velocity, radians(15)) for velocity in initialvelocity]
rangefromiv30 = [rangeformula(velocity, radians(30)) for velocity in initialvelocity]
rangefromiv45 = [rangeformula(velocity, radians(45)) for velocity in initialvelocity]
rangefromiv60 = [rangeformula(velocity, radians(60)) for velocity in initialvelocity]
rangefromiv75 = [rangeformula(velocity, radians(75)) for velocity in initialvelocity]
plt.plot(initialvelocity, rangefromiv15, label = "15 degrees")
plt.legend()
plt.plot(initialvelocity, rangefromiv30, label = "30 degrees")
plt.legend()
plt.plot(initialvelocity, rangefromiv45, label = "45 degrees")
plt.legend()
plt.plot(initialvelocity, rangefromiv60, label = "60 degrees")
plt.legend()
plt.plot(initialvelocity, rangefromiv75, label = "75 degrees")
plt.legend()
plt.title('Range as a function of initial velocity at various angles')
plt.xlabel('Initial Velocity (m/s)')
plt.ylabel('Range (meters)')
plt.show()
```
